```python
from konlpy.tag import Okt
import re

def split_noun_sentences(text):
    okt = Okt()
    sentences = text.replace(". ",".")
    sentences = re.sub(r'([^\n\s\.\?!]+[^\n\.\?!]*[\.\?!])', r'\1\n', sentences).strip().split("\n")
    
    result = []
    for sentence in sentences:
        if len(sentence) == 0:
            continue
        sentence_pos = okt.pos(sentence, stem=True)
        nouns = [word for word, pos in sentence_pos if pos == 'Noun']
        if len(nouns) == 1:
            continue
        result.append(' '.join(nouns) + '.')
        
    return result
text = "14일에 혜진 언니랑 엽떡 먹기"
split_noun_sentences(text)
```

```python
import re
from konlpy.tag import Okt
import re
from krwordrank.word import KRWordRank

def split_noun_sentences(text):
    okt = Okt()
    # sentences = text.replace(". ",".")
    # sentences = re.sub(r'([^\n\s\.\?!]+[^\n\.\?!]*[\.\?!])', r'\1\n', sentences).strip().split("\n")
    sentences = text.split('\n')
    result = []
    for sentence in sentences:
        sentence = sentence.strip()
        if len(sentence) == 0:
            continue
        sentence_pos = okt.pos(sentence, stem=True)
        nouns = [word for word, pos in sentence_pos if pos == 'Noun']
        if len(nouns) == 1:
            continue
        result.append(' '.join(nouns))
        
    return result

def extract_date_and_tasks(text):
    # 날짜 패턴 추출 (예: "14일에" "10/20(금)")
    date_pattern = r'(\d{1,2}일에|\d{1,2}/\d{1,2}\([^)]+\))'
    dates = re.findall(date_pattern, text)

    return dates

texts = split_noun_sentences(text)
text = "14일에 혜진 언니랑 엽떡 먹기\n10/20(금)에 자율 PJT 라이브 강의 이벤트 당첨자 안내드립니다"
dates = extract_date_and_tasks(text)
print(dates)
print(texts)
```
